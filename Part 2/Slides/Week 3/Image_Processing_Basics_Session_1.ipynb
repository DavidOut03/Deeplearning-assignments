# Activity II- BINARIZATION
## what is binarization? and why do we need binarization?

Binarization is the method of converting any gray – scale image (multi tone image) into black – white image (two tone image) . To perform binarization process, first find the threshold value of gray scale and check whether a pixel having a particular gray value or not

```python
plate=cv2.imread("/content/drive/MyDrive/Session1/gray_plate.jpg")
cv2_imshow(plate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Calling threshold funtion and setting threshold as 70, you can try different value(s) to see how it works
ret, thresh=cv2.threshold(plate, 70, 255, cv2.THRESH_BINARY)
# Display the Binary Image
cv2_imshow(thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Activity III- Histogram Equalization & CLAHE

```python
# For this activity, read the moon image from the directory, and set the path of the image
low_contrast=cv2.imread("/content/drive/MyDrive/Session1/moon.jpg")
cv2_imshow(low_contrast)
```

```python
gray = cv2.cvtColor(low_contrast, cv2.COLOR_BGR2GRAY)
equalized_img = cv2.equalizeHist(gray)
```

```python
plt.hist(gray.flat, bins=100, range=(0,255))
plt.hist(equalized_img.flat, bins=100, range=(0,255))
```

### Histogram Equalization result

```python
# diaplay the original image
cv2_imshow(gray)
# display the histogram equalized Image
cv2_imshow (equalized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Try CLAHE

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
equalized = clahe.apply(gray)
```

```python
#Original Image
cv2_imshow(gray)
# Eqalized Image
cv2_imshow (equalized_img)
# CLAHE
cv2_imshow (equalized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
